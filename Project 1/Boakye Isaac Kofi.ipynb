# Our dataset needs cleaning, meaning it is a bad dataset since it contains null values(NaN) and duplicates

# I will replaced all the null values with their mean values respectively

```python
mean_FA = dataset.fixed_acidity.mean()
mean_VA = dataset.volatile_acidity.mean()
mean_CA = dataset.citric_acid.mean()
mean_RS = dataset.residual_sugar.mean()
mean_C = dataset.chlorides.mean()
mean_FSD = dataset.free_sulfur_dioxide.mean()
mean_TSD = dataset.total_sulfur_dioxide.mean()
mean_D = dataset.density.mean()
mean_pH = dataset.pH.mean()
mean_S = dataset.sulphates.mean()
mean_A = dataset.alcohol.mean()
mean_Q = dataset.quality.mean()
```

```python
print(f'fixed_acidity: {mean_FA}')
print(f'volatile_acidity: {mean_VA}')
print(f'citric_acid: {mean_CA}')
print(f'residual_sugar: {mean_RS}')
print(f'chlorides: {mean_C}')
print(f'free_sulfur_dioxide: {mean_FSD}')
print(f'total_sulfur_dioxide: {mean_TSD}')
print(f'density: {mean_D}')
print(f'pH: {mean_pH}')
print(f'sulphates: {mean_S}')
print(f'alcohol: {mean_A}')
print(f'quality: {mean_Q}')
```

```python
dataset['fixed_acidity'] = dataset['fixed_acidity'].fillna(mean_FA)
dataset['volatile_acidity'] = dataset['volatile_acidity'].fillna(mean_VA)
dataset['citric_acid'] = dataset['citric_acid'].fillna(mean_CA)
dataset['residual_sugar'] = dataset['residual_sugar'].fillna(mean_RS)
dataset['chlorides'] = dataset['chlorides'].fillna(mean_C)
dataset['free_sulfur_dioxide'] = dataset['free_sulfur_dioxide'].fillna(mean_FSD)
dataset['pH'] = dataset['pH'].fillna(mean_pH)
dataset['sulphates'] = dataset['sulphates'].fillna(mean_S)
dataset['alcohol'] = dataset['alcohol'].fillna(mean_A)
dataset['quality'] = dataset['quality'].fillna(mean_Q)
```

# Matplotlib Correlation Heatmap

```python
import matplotlib.pyplot as plt
plt.imshow(dataset.corr(), cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(dataset.columns)), dataset.columns, rotation=90)
plt.yticks(range(len(dataset.columns)), dataset.columns)
plt.gcf().set_size_inches(14,12)

labels = dataset.corr().values
for y in range(labels.shape[0]):
    for x in range(labels.shape[1]):
        plt.text(x, y, '{:.2f}'.format(labels[y, x]), ha='center', va='center', color='white')
    
```
